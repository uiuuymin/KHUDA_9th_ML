```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, roc_auc_score
```

```python
# dataset repository path
url = "https://raw.githubusercontent.com/gaeng02/KHUDA-9th-ML/main/week4/dataset.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
### answer init
answer = [0] * 5
```

## Task 0. 초기 설정
다음 설정에 맞게 모델을 만드세요. <br>
별도의 정답은 없지만, 완료되지 않은 경우 Task 수행 불가 <br>
`X_train, y_train`을 이용하여 학습까지 진행하세요.

- feature(X) : `x1~x8`
- target(y) : `target`
- split : `test_size = 0.2, random_state = 212, stratify = y`
- model : `LogisticRegression(max_iter =2000, solver = "lbfgs")`

<br>

이때, X는 `StandardScaler`를 이용하여 데이터 전처리를 진행해주세요.

```python
### Step 1. feature, target Setting
feature_cols = [f"x{i}" for i in range (1, 9)]
X = df[feature_cols]
y = df["target"]

### Step 2. Data Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 212, stratify = y)

### Step 3. Data Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

### Step 4. Modeling
model = LogisticRegression(max_iter=2000, solver="lbfgs")
model.fit(X_train_scaled, y_train)

```

## Task 1
test dataset(`X_test`)을 이용하여 Accuracy를 구하세요.  
이때 Accuracy를 `answer[1]` 저장하세요.

```python
## Input Box
y_pred = model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)

answer[1] = accuracy
print(answer[1])
```

## Task 2
`threshold = 0.5`로 Label을 만드세요.  
그리고 `confusion_matrix`을 이용해 TP(True Positive) 개수를 구하세요.  
이때, TP 개수를 answer[2]에 저장하세요.

```python
## Input Box
y_prob = model.predict_proba(X_test_scaled)[:, 1]

y_pred = (y_prob >= 0.5).astype(int)

cm = confusion_matrix(y_test, y_pred)

tp = cm[1,1]

answer[2] = tp
print(answer[2])
```

## Task 3
이번에는 `threshold = 0.6`으로 F1 Score를 구하세요. <br>
이때, f1 score를 `answer[3]`에 저장하세요.

```python
## Input Box

y_prob = model.predict_proba(X_test_scaled)[:, 1]
y_pred = (y_prob >= 0.6).astype(int)

f1 = f1_score(y_test, y_pred)

answer[3] = f1
print(answer[3])
```

## Task 4
threshold를 0.05~0.95까지 0.01 간격으로 바꿔가며 최대가 되는 threshold를 찾으세요. <br>
best_threshold를 answer[5]에 저장하세요. <br>
(만약, F1 score가 동일하다면, 작은 threshold를 선택) <br>

```python
## Input Box

best_threshold = 0.0
best_f1 = 0.0

for threshold in np.arange(0.05, 0.96, 0.01) :
  y_pred = (y_prob >= threshold).astype(int)
  f1 = f1_score(y_test, y_pred)

  if f1 > best_f1 :
    best_f1 = f1
    best_threshold = threshold

answer[4] = round(best_threshold, 3)
print(answer[4])
```
